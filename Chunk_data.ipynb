```python
import pandas as pd
```

```python
df = pd.read_csv("users.6M0xxK.2024.public.csv")
```

```python
df.head()
```

```python
# Calculate the size of each chunk
chunk_size = len(df) // 5 + (1 if len(df) % 5 else 0)

# Split the dataframe into chunks and save them
for i in range(5):
    start = i * chunk_size
    end = start + chunk_size
    chunk_df = df.iloc[start:end]
    chunk_df.to_csv(f'chunk-data/chunk{i+1}.csv', index=False)

# Save each chunk as a CSV file in the 'chunk-data' folder.
# - For a **single file**, use `i+1` (chunk numbering starts from 1).
# - If **multiple files** are being processed, 
#   increment `i` based on the file order to prevent overwrites:
#   - First file: `i+1`
#   - Second file: `i+5`
#   - Third file: `i+10`
#   - (Continue increasing in steps of 5 for subsequent files)
```
